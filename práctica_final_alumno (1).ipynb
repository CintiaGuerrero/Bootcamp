```python
#A. Una función que calcule el precio total de la compra

def precio(articulos,compra):
  price = 0
  for index in compra:
    price += articulos[index]
  return price

precio(articulos, compra)
```

```python
#B. Una función que calcule el precio total de la compra con un descuento del 10%

def desc10(articulos,compra):
  price = 0
  for index in compra:
    price += articulos[index]
  return price*0.9

desc10(articulos,compra)
```

```python
#C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def desc10_mas100(articulos,compra):
  price = 0
  for index in compra:
    price += articulos[index]
  if price > 100:
    price*=0.9
  return price

desc10_mas100(articulos, compra)
```

```python
#D. Una función que calcule el precio total aplicando el IVA (21%)

def precioIVA(articulos,compra):
  price = 0
  for index in compra:
    price += articulos[index]
  price *= 1.21
  return round(price,2)

precioIVA(articulos,compra)
```

```python
#E.lista los artículos cuyo precio es superior a 20€

def art_mas20(articulos):
  mas20 = []
  for key, value in articulos.items():
    if value > 20:
      mas20.append(key)
  return mas20

art_mas20(articulos)
```
